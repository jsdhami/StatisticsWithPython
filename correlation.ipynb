# Correlation
- In statistics, correlation is a statistical relationship between two random variables or bivariate data.

```python
import pandas as pd

df = pd.read_csv('./data/tvObs.csv')
print(df)

# find the correlation between the variables
print("The correlation is\n", df.corr())

# plot graph of X[R1] vs Y[R2] using scatter plot and find the correlation between the variables
import matplotlib.pyplot as plt
import seaborn as sns

# plt.scatter(df['R1'], df['R2'])
sns.lmplot(x="R1", y="R2", data=df);
plt.title('Scatter plot of X[R1] vs Y[R2]')
plt.xlabel('Rank by X[R1]')
plt.ylabel('Rank by Y[R2]')


plt.show()

```

```python
# find mean, median, mode, range, variance, standard deviation, skewness, kurtosis for each column
print("Mean is\n", df.mean())
print("Median is\n", df.median())
print("Mode is\n", df.mode())
print("Range is\n", df.max()-df.min())
print("Variance is\n", df.var())
print("Standard deviation is\n", df.std())
print("Skewness is\n", df.skew())
print("Kurtosis is\n", df.kurt())
# quartiles and IQR
print("Q1 is\n", df.quantile(0.25))
print("Q2 is\n", df.quantile(0.5))
print("Q3 is\n", df.quantile(0.75))
print("IQR is\n", df.quantile(0.75) - df.quantile(0.25))
```
